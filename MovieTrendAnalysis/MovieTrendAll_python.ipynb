```python
import matplotlib.pyplot as plt
import pandas as pd

data_chart1 = {
    'Years': [2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022],
    'Box office revenue': [0.44,0.78,0.89,0.97,0.96,1.08,1.16,1.24,1.46,1.55,1.66,1.72,1.74,1.76,1.81,1.91,0.51,0.58,1.16]
}

df_chart1 = pd.DataFrame(data_chart1)

plt.plot(df_chart1['Years'], df_chart1['Box office revenue'], marker='o')
plt.title('Box Office Revenue in South Korea (2004-2022)')
plt.xlabel('Years')
plt.ylabel('Box Office Revenue (trillion South Korean Won)')
plt.grid(True)

plt.ylim(0, max(df_chart1['Box office revenue']) + 0.1)

plt.xticks(df_chart1['Years'])

plt.savefig('chart1_line_plot_modified.svg')
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio

data_chart3 = {
    'Years': [2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'South Korean films': [54.50, 57.80, 63.60, 49.90, 42.10, 48.70, 46.60, 51.90, 58.80, 59.70, 50.10, 52.00, 53.70, 51.80, 50.90, 51.00, 68.00, 30.10, 55.70],
    'Non-South Korean films': [45.50, 42.20, 36.40, 50.10, 57.90, 51.30, 53.40, 48.10, 41.20, 40.30, 49.90, 48.00, 46.30, 48.20, 49.10, 49.00, 32.00, 69.90, 44.30]
}

df = pd.DataFrame(data_chart3)

traces = []
trace1 = go.Scatter(
    x=df['Years'],
    y=df['South Korean films'],
    fill='tonexty',
    mode='lines',
    name='South Korean films',
)

trace2 = go.Scatter(
    x=df['Years'],
    y=df['Non-South Korean films'],
    fill='tonexty',
    mode='lines',
    name='Non-South Korean Films',
)

layout = go.Layout(
    title='Domestic vs. Foreign Films Market Share in South Korea (2004-2022)',
    xaxis=dict(title='Years', tickmode='linear'),
    yaxis=dict(title='Market Share Percentage', range=[0, 100]),
    showlegend=True
)

fig = go.Figure(data=[trace1, trace2], layout=layout)

svg_filename = 'chart3_stream_graph_plot.svg'
pio.write_image(fig, svg_filename)

fig.show()

print(f"The plot has been saved as {svg_filename}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

data_chart4 = {
    'Titles': ['The Roundup', 'Top Gun: Maverick', 'Avatar: The Way of Water', 'Hansan: Rising Dragon',
               'Confidential Assignment 2: International', 'Doctor Strange in the Multiverse of Madness',
               'Hunt', 'The Night Owl', 'Jurassic World Dominion', 'The Witch: Part 2. The Other One'],
    'Most watched movies in 2022 (in millions)': [12.69, 8.18, 7.31, 7.26, 6.98, 5.88, 4.35, 3.22, 2.84, 2.81],
    'Origin': ['South Korean', 'International', 'International', 'South Korean', 'South Korean', 'International',
               'South Korean', 'South Korean', 'International', 'South Korean']
}

df_chart4 = pd.DataFrame(data_chart4)

df_chart4_sorted = df_chart4.sort_values(by='Most watched movies in 2022 (in millions)', ascending=True)

colors = df_chart4_sorted['Origin'].map({'South Korean': '#436ACA', 'International': 'lightcoral'})

legend_patches = [
    mpatches.Patch(color='lightcoral', label='South Korean'),
    mpatches.Patch(color='#436ACA', label='International')
]

plt.barh(df_chart4_sorted['Titles'], df_chart4_sorted['Most watched movies in 2022 (in millions)'], color=colors)
plt.title('Most Watched Movies in South Korea (2022)')
plt.xlabel('Number of Viewers (millions)')
plt.grid(axis='x')

plt.legend(handles=legend_patches, loc='upper right')

plt.savefig('chart4_horizontal_bar_plot_sorted_colored_with_legend.svg')
plt.show()
```
